```python
# Convert Date column from object type to datetime

df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Add a "day of the week" column by extracting the weekday property of the dt property 
# of the Date column and place it into the 'day of the week' column

df["day of the week"] = df["Date"].dt.weekday
df[["Date","day of the week"]].head(15)
```
